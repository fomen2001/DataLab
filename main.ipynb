```python
import pandas as pd
import paths

def load_df(filepath=paths.location_path):
    print(f"File path : {filepath}")
    try:
        # On essaie d'accéder à la variable 'location_path' dans le module 'paths'
        file_path = filepath

        # Lecture du fichier CSV
        df_location = pd.read_csv(file_path)

    except AttributeError:
        print("Le chemin du fichier est introuvable.")
    except FileNotFoundError:
        print("Le fichier spécifié est introuvable.")
    return df_location

```

```python
from geopy.geocoders import Nominatim
def get_coordinates(address):
    geolocator = Nominatim(user_agent="geo_converter")
    location = geolocator.geocode(address)

    if location:
        return location.latitude, location.longitude
    else:
        return None
```

```python
# Define clean_address with handling for non-string values
def clean_address(address):
    if isinstance(address, str):  # Ensure the input is a string
        address = address.split(',')
        if len(address) > 3:
            address = address[-3:]
        return ','.join(address)  # Rejoin the cleaned parts
    else:
        return None  # Return None for invalid or missing
```

```python
new_df['clean_address'] = new_df['address_full'].apply(clean_address)
```

```python
import time

def get_coordinates(address):
    geolocator = Nominatim(user_agent="my_geocoder", timeout=10)  # Timeout de 10 secondes
    try:
        location = geolocator.geocode(address)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except Exception as e:
        print(f"Erreur pour l'adresse '{address}': {e}")
        return None, None
```

```python
new_df['latitude'], new_df['longitude'] = zip(*new_df['clean_address'].apply(get_coordinates))
```
